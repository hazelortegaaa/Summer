```python
import pandas as pd
import pandasql as ps #new library as sql so we call it ps
```

```python
data = pd.read_csv ('https://raw.githubusercontent.com/hazelortegaaa/data/main/Social_Network_Ads.csv')
```

```python
data.query('Age == 30') #makes all patient the same age
```

```python
data.query('Age > 30') #makes all patient GREATER than 30 age
```

```python
data.query('Age < 30') #makes all patients less than 30 years old
```

```python
data.query('Gender == "Female"') #You have to be mindful of the dataset
```

```python
#Using pandasql

query = "SELECT * FROM data WHERE Age == 30" #its going to hold a string, STAR  denotes everything 
ps.sqldf(query) #its going to bring everything from the data set, its passing it hrough the query ad sqldf
```

```python
#Using pandasql

query = "SELECT * FROM data WHERE GENDER == 'Male' AND  Age == 30" #its going to hold a string, STAR  denotes everything 
ps.sqldf(query) #its going to bring everything from the data set, its passing it hrough the query ad sqldf
```

```python
query = "SELECT * FROM data WHERE GENDER == 'Male' AND EstimatedSalary > 30" #its going to hold a string, STAR  denotes everything 
ps.sqldf(query) #its going to bring everything from the data set, its passing it hrough the query ad sqldf
```

```python
query = "SELECT * FROM data WHERE EstimatedSalary >= 130000" #its going to hold a string, STAR  denotes everything 
ps.sqldf(query) #its going to bring everything from the data set, its passing it hrough the query ad sqldf
```
